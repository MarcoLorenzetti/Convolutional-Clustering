## About the *Convolutional Clustering*

This study presents a novel approach to clustering two-feature data utilizing convolutional operations. By representing data points as matrices, convolutions are applied to identify dense regions, facilitating accurate assignment to corresponding clusters. This extension of traditional image processing techniques to diverse data domains leverages convolutions' ability to capture complex patterns and spatial relationships. My proposed algorithm overcomes limitations of distance-based clustering methods, providing improved accuracy in capturing intricate spatial dependencies. We implement and evaluate the algorithm on diverse datasets, comparing its performance to established techniques. This research demonstrates the efficacy and implications of convolution-based clustering approach, fostering advancements in data analysis and pattern recognition.

```python
# Generazione delle coordinate x e y con distribuzione gaussiana
num_punti = 600
x = [np.random.choice([-50, 0, 50])+np.random.normal(30, 10) for x in range(num_punti)]
y = [np.random.choice([-5, 50, -40])+np.random.normal(30, 8) for x in range(num_punti)]

# Creazione del dataframe
df = pd.DataFrame({'x': x, 'y': y})
```

```python
# # Creazione del dataframe di esempio
# num=1000
# df = pd.DataFrame({'x': np.random.randint(-100, 201, size=num),
#                    'y': np.random.randint(-100, 201, size=num)})
# # df = pd.read_csv('clustering data/shopmania.csv')
# # df = df.iloc[:, 1:]
plt.figure(figsize=(6,6))
df.plot(df.columns[0], df.columns[1], kind='scatter')
df.head()
```

```python
for i in range(1,4):
    print(f'-------- Iterations (loops): {i} --------')
    df2 = CC(df,
             scale_size=800, pad=None, plot_matrix=False,
             kernel_type='gaussian', k_size=None, sigma=40,
             ratio_scaler=0.9, THR=50, loops=i,
             draw_type='line', radius=2, thickness=1,
             show_clusters=True)


    plt.figure(figsize=(6,6))
    plt.scatter(df2[df2.columns[0]], df2[df2.columns[1]], c=df2['label']);
    plt.show()
```
